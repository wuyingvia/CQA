```python
def remove_moving_window(speed):

    batch_num, n_pred, n_node = speed.shape
    select_idx = np.arange(0, batch_num, n_pred)
    speed = speed[select_idx, :, :]

    speed = np.reshape(speed, (-1, n_node))
    print(speed.shape)

    return speed
```

visualization of prediction interval

```python
plot_day_len = 1
plot_start =  156 
plot_sensor_id = 66


fontsize = 14
x_ticks_pos = np.linspace(0, 288, 5)
x_tick_labels = ['18:00','00:00', '06:00', '12:00','18:00']

fig, axs = plt.subplots(2, 4, sharex=True, sharey=True, figsize=(20, 10))
plt.cla()
plt.subplots_adjust(wspace=0.18)
plt.subplots_adjust(hspace=0.2)

x_range = np.arange(plot_day_len * 288)


handle_ground = []
handle_outline = []
handle_interval = []


def plot_subfig(ax, groundtruth, lower, upper, name):

    # plot ground truth
    plot_groundtruth = groundtruth[plot_start : plot_day_len*288 + plot_start, plot_sensor_id]
    handle_ground = ax.plot(plot_groundtruth, '.', color = 'black', label='GroundTruth')

    # find outliner
    out_idx = np.where(
        (plot_groundtruth < lower[plot_start : plot_day_len*288 + plot_start, plot_sensor_id] ) | (
        upper[plot_start : plot_day_len*288 + plot_start, plot_sensor_id] < plot_groundtruth)
        )
    handle_outline = ax.plot(x_range[out_idx], plot_groundtruth[out_idx],'.', color = 'red', label='GroundTruth')

    ax.plot(lower[plot_start : plot_day_len*288 + plot_start, plot_sensor_id],linestyle = '-', color = 'grey')
    ax.plot(upper[plot_start : plot_day_len*288 + plot_start, plot_sensor_id],linestyle = '-', color = 'grey')
    handle_interval = ax.fill_between(
        x_range,
        lower[plot_start : plot_day_len*288 + plot_start, plot_sensor_id],
        upper[plot_start : plot_day_len*288 + plot_start, plot_sensor_id],
        facecolor = 'grey', alpha = 0.5
    )

    ax.set_xlim([x_range[0], x_range[-1]])
    ax.set_xticks(x_ticks_pos)
    ax.set_xticklabels(x_tick_labels)

    ax.set_title('{:s} \n Coverage:{:.2f}%  Interval:{:.2f}'.format(
        name, 
        100 - len(out_idx[0])/plot_day_len/288 * 100, 
        np.mean(
            upper[plot_start : plot_day_len*288 + plot_start, plot_sensor_id] - 
            lower[plot_start : plot_day_len*288 + plot_start, plot_sensor_id]
        )), fontsize=fontsize)
    ax.tick_params(labelsize=fontsize, colors='k')

    ax.grid()

plot_subfig(axs[0,0], dropout_groundtruth,  ha_upper, ha_lower,'ha-w')
plot_subfig(axs[0,1], dropout_groundtruth, dropout_lower * (dropout_lower>0) , dropout_upper, 'dropout')
plot_subfig(axs[0,2], bayesian_groundtruth, bayesian_lower, bayesian_upper, 'bayesian')
plot_subfig(axs[0,3], conformal_groundtruth, conformal_lower, conformal_upper, 'conformal')
plot_subfig(axs[1,0], quantile_groundtruth, quantile_lower, quantile_upper, 'quantile')
plot_subfig(axs[1,1], quantile_groundtruth, quantile_conformal_lower, quantile_conformal_upper, 'quantile conformal')
plot_subfig(axs[1,2], adaptive_groundtruth, adaptive_lower, adaptive_upper,  'adaptive')



[axs[1,i].set_xlabel('Time', fontsize=fontsize) for i in range(4)]
[axs[i,0].set_ylabel('Flow', fontsize=fontsize) for i in range(2)]

plt.delaxes(axs[1,3])

fig.legend([handle_ground, handle_outline, handle_interval], 
            labels=['groundtruth', 'out of coverage', 'interval'],
           loc='center',  bbox_to_anchor=(0.72, 0.25), fontsize=fontsize+4, 
framealpha=0.5, borderaxespad=0,  ncol=1)
# ax.grid()

# ax.legend(fontsize=fontsize)

# fig.show()

plt.savefig('interval.pdf', bbox_inches='tight')
```

### coverage and interval by nodes

```python
ontsize = 14
x_ticks_pos = np.linspace(0, 288, 5)


# fig, axs = plt.subplots(2, 1, sharex=True, figsize=(16, 7))
# plt.cla()
# # plt.subplots_adjust(wspace=0.18)
# plt.subplots_adjust(hspace=0.1)

fig = plt.figure(figsize=(24, 10))

gs = fig.add_gridspec(2, 3, width_ratios=(14,1.5,1.5), height_ratios=(1,1), wspace=0.05)





def plot_node_coverage_interval(axs, groundtruth, lower, upper, name):
    
    node_interval = np.mean(upper - lower, axis=0)
    is_covered = ((lower <= groundtruth) & (upper >= groundtruth))
    node_coverage = np.sum(is_covered, axis=0) / groundtruth.shape[0]*100

    if name == 'quantile conformal':
        axs[0].plot(node_coverage, '.', label=name)
        axs[1].plot(node_interval, '.')
    elif name == 'adaptive':
        axs[0].plot(node_coverage, '^', label=name)
        axs[1].plot(node_interval, '^')
    else:
        pass
    return node_coverage, node_interval


axs_cv = fig.add_subplot(gs[0,0])
axs_it = fig.add_subplot(gs[1,0])


qc_cv, qc_it = plot_node_coverage_interval([axs_cv, axs_it], quantile_conformal_groundtruth, quantile_conformal_lower, quantile_conformal_upper, 'quantile conformal')
ada_cv, ada_it = plot_node_coverage_interval([axs_cv, axs_it], adaptive_groundtruth, adaptive_lower, adaptive_upper, 'adaptive')

axs_cv.set_ylabel('(a) Coverage (%)', fontsize = fontsize)
axs_cv.tick_params(labelsize=fontsize)

axs_it.set_ylabel('(b) Interval (veh/l/5min)', fontsize = fontsize)
axs_it.set_xlabel('Node No.', fontsize = fontsize)
axs_it.tick_params(labelsize=fontsize)

fig.legend(loc='center',  bbox_to_anchor=(0.5, 0.92), fontsize=fontsize+4, 
framealpha=0.5, borderaxespad=0,  ncol=5)


axs_cv_1 = fig.add_subplot(gs[0,1], sharey=axs_cv)
axs_cv_2 = fig.add_subplot(gs[0,2], sharey=axs_cv, sharex=axs_cv_1)

bins = np.arange(50, 100, 1)
axs_cv_1.hist(qc_cv, bins=bins, color=plt.cm.tab10(0), orientation='horizontal')
axs_cv_2.hist(ada_cv, bins=bins, color=plt.cm.tab10(1), orientation='horizontal')

axs_it_1 = fig.add_subplot(gs[1,1], sharey=axs_it)
axs_it_2 = fig.add_subplot(gs[1,2], sharey=axs_it, sharex=axs_it_1)

bins = np.arange(20, 210, 5)
axs_it_1.hist(qc_it, bins=bins, color=plt.cm.tab10(0), orientation='horizontal')
axs_it_2.hist(ada_it, bins=bins, color=plt.cm.tab10(1), orientation='horizontal')



for t in [axs_cv_1, axs_cv_2, axs_it_1, axs_it_2]:
    # t.axis('off')
    t.tick_params(axis='x', labelbottom=False)
    t.tick_params(axis='y', labelleft=False)


# fig.show()
fig.savefig('nodeqca_2.pdf', bbox_inches='tight')
```
